### Boxplot

```python
# altair boxplot example
class1 = df[df.Pclass == 1]

alt.Chart(class1).mark_boxplot(size = 100).encode(
    alt.X('Pclass:O'), alt.Y('Fare', scale=alt.Scale(domain=(0, 500))),
    tooltip = ["Fare"]).properties(width = 300, height = 300,
                                   title = 'Distribution of Fares for Class 1')
```

### Barplot

```python
# altair example
alt.Chart(df).mark_bar().encode(alt.X('Pclass', type = 'ordinal'), 
                                                  alt.Y('count()', axis = alt.Axis(title = 'Number of Passengers')
                                                  ), color = 'Survived:N').properties(width = 300, height = 300)
```

Note: You can substitute 'Pclass' for other passenger categories (for example, 'Sex' or 'SibSp').
